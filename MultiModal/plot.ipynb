```python
import numpy as np
import matplotlib.pyplot as plt

def read_list(fp):
    file = open(fp, 'r')

    convert = lambda s : (list(map(int, )))
    s = file.readline()
    x = s[1:-2].split(',')
    x = list(map(float, x))

    return x

SingleLSTM = read_list('log.SingleLSTM.txt')
SingleBERT = read_list('log.SingleBERT.txt')
BERTMulMod = read_list('log.BERTMulMod.txt')
BERTMulModMulTask = read_list('log.BERTMulModMulTask.txt')

l = min([len(SingleBERT), len(SingleLSTM), len(BERTMulMod), len(BERTMulModMulTask)])
[len(SingleBERT), len(SingleLSTM), len(BERTMulMod), len(BERTMulModMulTask)]
```

```python
def average(lst):
    s = 0.0
    for i in range(len(lst)):
        s += lst[i]
        lst[i] = s
    
    for i in range(len(lst)):
        lst[i] /= i+1
    
    return lst

SingleLSTM = average(SingleLSTM[:l])
SingleBERT = average(SingleBERT[:l])
BERTMulMod = average(BERTMulMod[:l])
BERTMulModMulTask = average(BERTMulModMulTask[:l])

x = np.array([i for i in range(l)])
```

```python
fig, ax = plt.subplots()

l1 = ax.plot(x, SingleLSTM, label='SingleLSTM')
l2 = ax.plot(x, SingleBERT, label='SingleBERT')
l3 = ax.plot(x, BERTMulMod, label='BERTMulMod')
l4 = ax.plot(x, BERTMulModMulTask, label='BERTMulModMulTask')

ax.set_title('Training Loss')
ax.set_xlabel('Batch')
ax.set_ylabel('Loss')

ax.legend()
plt.show()
```
